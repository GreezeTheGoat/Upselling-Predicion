```python
import pandas as pd
from pathlib import Path
import yaml
import sys
import yaml
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
```

```python
path_yaml = Path("../src/config.yaml")
try:

    with open(path_yaml, "r") as file:
        config = yaml.safe_load(file)

except FileNotFoundError:
    print("Config file not found")
    sys.exit(1)
```

```python
path_raw = Path(config["paths"]["raw"])
path_processed = Path(config["paths"]["processed"])

df = pd.read_csv(path_raw)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('future.no_silent_downcasting', True)
```

```python
df = df[df["Churn Value"] == 0]
```

```python
useless_columns = config["features"]["useless_columns"]
```

```python
df = df.drop(columns=useless_columns)
```

```python
raw_standardization = config["data_standardization"]
map_to_apply = {}
for std_key, raw_values in config["data_standardization"].items():
    if isinstance(raw_values, list):
        for val in raw_values:
            map_to_apply[val] = std_key
    else:
        map_to_apply[raw_values] = std_key

df = df.replace(map_to_apply)
```

```python
for col in config["features"]["nominal_categoricals"]:
    df[col] = df[col].astype(str).str.lower().str.strip()

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), config["features"]["numerical_features"]),
        ("ord", OrdinalEncoder(categories=config["features"]["ordinal_orders"]), config["features"]["ordinal_categoricals"]),
        ("nom", OneHotEncoder(drop="first", handle_unknown="ignore", sparse_output=False), config["features"]["nominal_categoricals"]),
    ],
    verbose_feature_names_out=False,
    remainder= "passthrough"
    )

preprocessor.set_output(transform="pandas")

df = preprocessor.fit_transform(df)
```

```python
df.columns = [
    col.replace("_yes", "") if col.endswith("_yes") else col 
    for col in df.columns
]
```

```python
extra_services_columns = config["features"]["extra_services"]
df["Total Extra Services"] = df[extra_services_columns].sum(axis = 1)
```

```python
df["is_fiber"] = (df["Internet Service"] == 2).astype(int)
```

```python
df.to_csv(path_processed.joinpath("telco_preprocessed.csv"), index=False)
```
